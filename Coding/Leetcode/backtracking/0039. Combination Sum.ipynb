```python
'''
39. Combination Sum
'''
class Solution:
    def combinationSum(self, candidates: List[int], target: int) -> List[List[int]]:
        self.result = []
        self.backtracking(candidates, [], target, 0, 0)
        return self.result

    def backtracking(self, candidates, path, target, sum_, startIndex):
        if sum_ > target: return
        if sum_ == target:
            self.result.append(path[:])
            return
        for i in range(startIndex, len(candidates)):
            path.append(candidates[i])
            sum_ += candidates[i]
            self.backtracking(candidates, path, target, sum_, i)
            sum_ -= candidates[i]
            path.pop()
        return

class Solution:
    def combinationSum(self, candidates: List[int], target: int) -> List[List[int]]:
        candidates.sort()
        self.result = []
        self.backtracking(candidates, [], target, 0, 0)
        return self.result

    def backtracking(self, candidates, path, target, sum_, startIndex):
        if sum_ > target: return
        if sum_ == target:
            self.result.append(path[:])
            return
        for i in range(startIndex, len(candidates)):
            if candidates[i] > target:
                break
            self.backtracking(candidates, path + [candidates[i]], target, sum_ + candidates[i], i)
        return


class Solution:
    def combinationSum(self, candidates: List[int], target: int) -> List[List[int]]:
        self.result = []
        max_size = target // min(candidates)
        for k in range(1, max_size + 1):
            self.backtracking(candidates, k, [], target, 0, 0)
        return self.result

    def backtracking(self, candidates, k, path, target, sum_, startIndex):
        if len(path) > k or sum_ > target: return
        if len(path) == k and sum_ == target:
            self.result.append(path[:])
            return
        for i in range(startIndex, len(candidates)):
            path.append(candidates[i])
            sum_ += candidates[i]
            self.backtracking(candidates, k, path, target, sum_, i)
            sum_ -= candidates[i]
            path.pop()
        return
```
